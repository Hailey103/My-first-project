```python
import pandas as pd
import numpy as np

#load the first dataset
dataset_1 = pd.read_csv('rental_bike_descr.csv')

#check the type of dataset
print(type(dataset_1))

#check its shape
print('Shape is', dataset_1.shape)

#print first five rows
print(dataset_1.head())

#load the second dataset
dataset_2 = pd.read_csv('rental_bike_season.csv')
print(dataset_2.head())
print(dataset_2.shape)
# drop columns
dataset_2 =dataset_2.drop(['Unnamed: 0'],axis=1)
print(dataset_2.head())
print(dataset_2.shape)

#merge these datasets 1 and 2
combined_data = pd.merge(dataset_1,dataset_2,on='instant')
print(combined_data)
combined_data.head()

#load the third dataset
dataset_3 = pd.read_csv('final_rental_bike_dataset.csv')
print(dataset_3.shape)

dataset_3.head()
dataset_3.tail(15)

#sort values of column
dataset_3 = dataset_3.sort_values(by =['instant'])
#check the products
dataset_3.tail(15)
dataset_3.head()

#concatenate the combined_data and dataset_3
final_data = pd.concat([combined_data,dataset_3])
final_data.shape

#rename and display the final_data
final_data = final_data.rename(columns ={'dteday':'date', 'yr':'year',  'mnth':'month',  'hr':'hour',
                                         'weathersit':'weather','hum':'humidity','cnt':'count'})
final_data.head()
# check for null values in final_data
final_data.isna()

#show the 11 null values in atemp
final_data.isna().sum(axis=0)

#Calculate the % of missing value
percentage_missing_values =(final_data['atemp'].isna().sum(axis=0)/final_data.shape[0])*100
percentage_missing_values

#drop null values by function dropna
final_data = final_data.dropna(axis=0)
final_data.shape

#check again, the 11 null values in atemp
final_data.isna().sum(axis=0)

#check if casual+registered = count
np.sum(final_data['casual']+final_data['registered']-final_data['count'])

#check month should be in the range of 1-12 months by the function of unique()
np.unique(final_data.month)
np.unique(final_data.hour)

#check the stastical summary of dataset
print(final_data.describe())

#Export the file to csv named consolidation
final_data.to_csv(r'consolidation.csv')
```
